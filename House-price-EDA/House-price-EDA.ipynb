# Graphical EDA

```python
import seaborn as sns
```

### Histogram

```python
sns.displot(x='YearBuilt', kde=True, data=data)
```

```python
sns.histplot(x='YearBuilt', y='SalePrice', data=data)
```

### Box plot

```python
sns.boxplot(x='YearBuilt', data=data)
```

```python
sns.boxplot(x='YearBuilt', y='SaleCondition', data=data)
```

### Violin Plot

```python
sns.violinplot(x='SalePrice', data=data)
```

```python
sns.violinplot(x='YearBuilt', y='CentralAir', data=data)
```

### Bar plot

```python
sns.countplot(x='SaleCondition', data=data)
```

```python
sns.countplot(x='SaleCondition', hue='CentralAir', data=data)
```

### Line plot

```python
sns.lineplot(x=data.loc[0:100].index, y='MSSubClass', data=data.loc[0:100])
```

### Scatter plot

```python
sns.scatterplot(x="LotFrontage", y="SalePrice", data=data)
```

```python
sns.scatterplot(x="LotFrontage", y="SalePrice", hue="CentralAir", data=data)
```

### Bubble plot

```python
sns.scatterplot(x="LotFrontage", y="SalePrice", data=data, size="YearBuilt")
```

### Pair plot

```python
sns.pairplot(data=data[['LotFrontage', 'SalePrice', 'YearBuilt']])
```

```python
sns.pairplot(data=data[['LotFrontage', 'SalePrice', 'YearBuilt', 'CentralAir']], hue='CentralAir')
```

### Joint Plot

```python
sns.jointplot(x='SalePrice', y='YearBuilt', data=data)
```

```python
sns.jointplot(x='SalePrice', y='YearBuilt', data=data, hue='CentralAir')
```

### Heat map

```python
heat_data = pd.pivot_table(data, index='MoSold', columns='YrSold', aggfunc=np.count_nonzero)['SalePrice'].fillna(0)
heat_data
```

```python
sns.heatmap(heat_data)
```
